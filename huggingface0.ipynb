```python
from transformers import pipeline
```

```python
classifier= pipeline('sentiment-analysis')
```

```python
classifier("I hate that movie and the cat is bad")
```

```python
classifier2= pipeline('sentiment-analysis', model= 'nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
classifier2("I love that film and the cast was good")
```

```python
results=classifier(["we are happy you are here","we hope you don't hate it"])
```

```python
for i in results:
  print(f"{i['label']} ,{round(i['score'],2)}")
```

5.Question Answering

```python
question_answering=pipeline("question-answering")
```

```python
context=r"""Extractive Question Answering is the task of extracting answer from from a text given a question"""
result=question_answering(question="what is the question answering?", context=context)
print(f"Answer: {result['answer']}, score: {round(result['score'],4)}")
```
